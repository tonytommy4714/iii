```python
import requests
import html5lib
from bs4 import BeautifulSoup
from openpyxl import Workbook
import time
import pandas as pd
import numpy as np
```

```python

head={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.139 Safari/537.36'}


i=20
table=[]
while i<=60: 
    line='https://www.591.com.tw/webService-market-1.html?firstRow='
    page=str(i)
    last='&totalRows=894&type=1&sectionid=5&kind=1&year=2016&orderType=desc&orderField=refreshtime'
    real_link=line+page+last
    res = requests.get( real_link , headers = head)
    print( real_link)
    soup = BeautifulSoup(res.text,"html.parser")
    df = pd.read_html(res.text)
    table.append(df[0].drop([0,1]))
    
    
    

   
   

       
       
    i+=20
    time.sleep(2)
jointable=pd.concat(table,ignore_index=True)#合併table重新給索引值
jointable.columns=['成交時間','用途','形態','街道','坪數','租金(車位)','格局','樓層']#給需要的欄位
jointable['樓']=np.nan #增加一個樓的新欄位
jointable['總樓層']=np.nan#增加一個總樓層的新欄位
i=0
while i<len(jointable):
    jointable.ix[i,['街道']]='大安區'+jointable.ix[i,['街道']]#把街道的每個值加上區域
    
    size=str(jointable.ix[i,['坪數']]).split()[1].replace('坪','')#選取坪數的值轉為字串後做split選取第一個元素替評為空白
    jointable.ix[i,['坪數']]=size #把更新後的值重新指定到原位置 
    
    clear_price=str(jointable.ix[i,['租金(車位)']]).split()[1].replace('元','')
    jointable.ix[i,['租金(車位)']]=clear_price
    
    if(str(jointable.ix[i,['樓層']]).split()[1].split('/')[0]=='+1'):
        jointable.ix[i,['樓']]=str(jointable.ix[i,['樓層']]).split()[1].split('/')[0][1]
    else:
        jointable.ix[i,['樓']]=str(jointable.ix[i,['樓層']]).split()[1].split('/')[0]
    jointable.ix[i,['總樓層']]=str(jointable.ix[i,['樓層']]).split()[1].split('/')[1]
    i+=1
jointable=jointable.drop(['樓層'],axis=1)
```
