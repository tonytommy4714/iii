```python
#抓總比數
def totalrow(a):
    head={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.139 Safari/537.36'}
    real_link=a
    res = requests.get( real_link , headers = head)
    soup = BeautifulSoup(res.text,"html.parser")
    page_number=soup.findAll('span',{'style':'float:right'})
    b=int(page_number[-1].get_text().replace('共','').replace('筆',''))
    return b 
print(totalrow('https://www.591.com.tw/webService-market-1.html?totalRows=&type=1&sectionid=3&kind=2&year=2018&orderType=desc&orderField=refreshtime'))
```
