```python
# 簡單的換頁爬蟲
url_template = "https://www.ptt.cc/bbs/watch/index{}.html"

start_page_no = 1648 #開始頁面數
pages_to_crawl = 10  #想要爬的頁數


for page_no in range(start_page_no, start_page_no - pages_to_crawl, -1): #製造一個數列，並且for loop
    url = url_template.format(page_no) #將page_no帶進去url template
    print(url)
    resp = requests.get(url)
    print(resp.status_code)
    with open('/tmp/ptt_watch_{}.html'.format(page_no), 'w') as f:
        f.write(resp.text)
```

# HTML Parsing

Beautifulsoup

https://www.crummy.com/software/BeautifulSoup/bs4/doc/

```python
!pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup
```

```python
house_dict = {}
```

```python
html = open('/tmp/xinyi.html').read()
```

```python
soup = BeautifulSoup(html, 'html5lib')
```

```python
title_tag = soup.select_one('#content-main > div > h2')
```

```python
title_tag
```

```python
house_dict['title'] = title_tag.text.strip()
house_dict
```

```python
house_dict['price'] = int(soup.select_one('#obj-info > ul > li > span').text.replace(',',''))
house_dict
```

```python
house_dict['price'] = int(soup.select('#obj-info > ul > li')[0].select_one('span').text.replace(',',''))
```

```python
house_dict['area'] = float(soup.select('#obj-info > ul > li')[1].text.replace("建物登記：", '').replace("坪", ''))
```

```python
house_dict['price_per_area'] = float(soup.select('#obj-info > ul > li')[2].text.replace("每坪單價：", '').replace("萬/坪", ''))
```

```python
house_dict
```

```python
import json
json.dump(house_dict, open('/tmp/houses_dict.json', 'w'), ensure_ascii=False)
```
