```python
from bs4 import BeautifulSoup
```

```python
myhtml = """<html>
<h1>CB101</h1>
<div cool-id="1">cool story bro</div>
<div cool-id="2">cool story bro</div>
<div cool-id="3">cool story bro</div>
<div cool-id="4">cool story bro</div>
<div cool-id="5">cool story bro</div>
<div cool-id="6">cool story bro</div>
<p>hi</p>
</html>"""
```

```python
soup = BeautifulSoup(myhtml, 'lxml')
```

```python
def my_search_func(ele):
    if ele.get('cool-id'):
        int_id = int(ele.get('cool-id'))
        if int_id < 4:
            return True
    else:
        return False
```

```python
soup.findAll(my_search_func)
```

# 列表頁爬蟲技巧

```python
import re

import requests

url = "https://www.ptt.cc/bbs/gossiping/index.html"

resp = requests.get(url, headers={"cookie": "over18=1"})
re.findall('''<a href="(/bbs/Gossiping/.+html)">''', resp.text)
```

```python
re.findall('''<a href="/bbs/Gossiping/M.1528178090.A.F45.html">''', resp.text)
```

```python
"""
.     -> 萬用字元
\d    -> 0-9
\w    -> a-zA-Z
\W    -> \n \t


{2}   -> repeat 2 times
{1,9} -> repeat min 1 time, max 9 times
+ -> greedy repeat 1-N times
* -> greedy repeat 0-N times
+? -> lazy repeat 1-N times
*? -> lazy repeat 0-N times
?

() -> capture group
"""
```

```python
re.findall('''<a href="/bbs/Gossiping/M.1528178090.A.F4\d.html">''', resp.text)
```

```python
re.findall('''<a href="/bbs/Gossiping/M.1528178090.A.F\d\d.html">''', resp.text)
```

```python
re.findall('''<a href="/bbs/Gossiping/M.1528178090.A.F\d{2}.html">''', resp.text)
```

```python
re.findall('''<a href="/bbs/Gossiping/M.1528178090.A.F\d{1,2}.html">''', resp.text)
```

```python
re.findall('''<a href="/bbs/Gossiping/M.1528178090.A.F\d+.html">''', resp.text)
```

```python
tt = '<a href="/bbs/Gossiping/M.1528178090.A.F123123123123123.html">23412341234.html">'

re.findall('''<a href="/bbs/Gossiping/M.1528178090.A.F.+?html''', tt)
```

```python
re.findall('''<a href="(/bbs/Gossiping/.+html)">''', resp.text)
```
