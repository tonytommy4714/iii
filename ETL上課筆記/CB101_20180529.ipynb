# 蔡英文就職演說

https://www.president.gov.tw/NEWS/20444

```python
import jieba
speech = open('/tmp/speech.txt').read()
jieba.set_dictionary('/Users/ian/Downloads/dict.txt.big')
```

```python
words = [word for word in jieba.cut(speech)]
```

```python
words
```

```python
stop_words = ['，', '。', '\n', '、', '的', '與']
res_dict = {}
for word in words:
    if word in stop_words:
        continue
    if word in res_dict:
        res_dict[word] += 1
    else:
        res_dict[word] = 1
```

```python
res_dict
```

```python
for key,value in res_dict.items():
    print(key)
    print(value)
```

```python
res_list = [[k,v] for k,v in res_dict.items()]
```

```python
res_list
```

```python
res_list = sorted(res_list, key=lambda x: x[1], reverse=True)
res_list
```

```python
import csv
```

```python
with open('/tmp/word_cnt.csv', 'w') as csvfile:
    dw = csv.writer(csvfile)
    dw.writerow(['word', 'count'])
    for row in res_list:
        dw.writerow(row)
```
